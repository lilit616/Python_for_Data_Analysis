Seaborn, a Python data visualization library, to create bar charts for statistical analysis

Seaborn is a Python data visualization library that provides simple code to create elegant visualizations for statistical exploration and insight. Seaborn is based on Matplotlib, but improves on Matplotlib in several ways:

1) Seaborn provides a more visually appealing plotting style and concise syntax.
2)Seaborn natively understands Pandas DataFrames, making it easier to plot data directly from CSVs.
3)Seaborn can easily summarize Pandas DataFrames with many rows of data into aggregated charts.

```
#load iris dataset
#1 - Swarm plots
iris = sns.load_dataset("iris")
sns.swarmplot(x="species",y="petal_length",data=iris)
plt.show()
```

```
#box plots
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
```

```
sns.set_style("whitegrid")
sns.boxplot(data=data)
plt.title("whitegrid")
plt.show()

sns.set_style("darkgrid")
sns.boxplot(data=data);
plt.title("darkgrid")
plt.show()
```

Seaborn Factor plots is a categorical plot. It lets us see what the data looks like when segmented by one or more variables.

```
#seaborn factor plots
titanic = sns.load_dataset("titanic")
g= sns.factorplot("class","survived","sex",data=titanic,kind="bar",palette="muted",legend=True)
plt.show()
```

```
#a simple colormap - bringing our seaborn colors into matplotlib
N = 500
current_palette = sns.color_palette("muted",n_colors=5)
cmap = ListedColormap(sns.color_palette(current_palette).as_hex())

data1 = np.random.randn(N)
data2 = np.random.randn(N)

colors = np.random.randint(0,5,N)

plt.scatter(data1,data2,c=colors,cmap=cmap)

plt.colorbar()

plt.show()
```

Two core functions in seaborn Axes level functions - regplot, boxplot, kdeplot. Figure level functions - Implot, factorplot, jointplot.

Figure Level - Are for exploratory analysis, they set up the matplot lib figure containing the plot and spread out the visualization across multiple axes.

Axes level - draw out onto the single matplotlib axes and dont affect otherwise the rest of the figure.

```
sns.set_context("paper", font_scale=1, rc={"font.size":2,"axes.labelsize":1})
```

```
#Implot - Statistical estimation ; Linear regression
tips = sns.load_dataset("tips")

sns.lmplot(x="total_bill", y="tip", col="time", hue="smoker",
           data=tips);
```

```
#catplot - to visualize categorical data, one numeric variable and one categorical variable.
sns.catplot(x="day", y="total_bill", hue="smoker",
            kind="violin", split=True, data=tips);
# the below is kernel density
```

```
#if you want the scatter

sns.catplot(x="day", y="total_bill", hue="smoker",
            kind="swarm", data=tips);
```
